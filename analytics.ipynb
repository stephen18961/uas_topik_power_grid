# Western States Power Grid
Used in the research of D. J. Watts and S. H. Strogatz, "Collective dynamics of `small-world' networks", Nature 393, 440-442 (1998). 


Power transmission lines as edges and 

Power stations as nodes

```python
import networkx as nx

# Load the graph
file_path = "data/power.gml"
G = nx.read_gml(file_path, label=None)

# Degree Centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)

# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness Centrality:", betweenness_centrality)

# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
print("Closeness Centrality:", closeness_centrality)

# Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)
print("Eigenvector Centrality:", eigenvector_centrality)

# PageRank
pagerank = nx.pagerank(G)
print("PageRank:", pagerank)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph
file_path = "data/power.gml"
G = nx.read_gml(file_path, label=None)

# Calculate centralities
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
pagerank = nx.pagerank(G)

# Function to get the top 10 nodes for a given centrality
def get_top_10(centrality):
    return sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:10]

# Get top 10 nodes for each centrality
top_10_degree = get_top_10(degree_centrality)
top_10_betweenness = get_top_10(betweenness_centrality)
top_10_closeness = get_top_10(closeness_centrality)
top_10_eigenvector = get_top_10(eigenvector_centrality)
top_10_pagerank = get_top_10(pagerank)

# Prepare data for histogram visualization
centrality_names = ["Degree", "Betweenness", "Closeness", "Eigenvector", "PageRank"]
centrality_values = [
    [value for _, value in top_10_degree],
    [value for _, value in top_10_betweenness],
    [value for _, value in top_10_closeness],
    [value for _, value in top_10_eigenvector],
    [value for _, value in top_10_pagerank],
]

# Plot histograms
fig, axes = plt.subplots(1, 5, figsize=(20, 5), sharey=True)
for ax, name, values in zip(axes, centrality_names, centrality_values):
    ax.bar(range(1, 11), values, color="skyblue")
    ax.set_title(name + " Centrality")
    ax.set_xlabel("Top 10 Nodes")
    ax.set_ylabel("Centrality Value")
    ax.set_xticks(range(1, 11))

plt.tight_layout()
plt.show()

# Output the top 10 nodes for reference
(top_10_degree, top_10_betweenness, top_10_closeness, top_10_eigenvector, top_10_pagerank)
```

## 1. Degree Centrality

    Interpretasi: Tingginya degree centrality menunjukkan bahwa node memiliki banyak koneksi langsung.
    
    Temuan: Node dengan degree centrality tertinggi (ID 2553 dan 4458) memiliki peran penting dalam menjaga konektivitas lokal jaringan. Meski demikian, nilai rata-rata degree cukup rendah, menunjukkan sifat jaringan yang sparsely connected.

## 2. Betweenness Centrality

    Interpretasi: Betweenness centrality tinggi menunjukkan bahwa node tersebut menjadi perantara penting dalam aliran informasi atau daya antara node lainnya.
    
    Temuan: Node seperti 4164, 2543, dan 1243 memiliki betweenness centrality yang sangat tinggi, menandakan perannya sebagai hub strategis untuk jalur daya. Jika node ini gagal, jaringan dapat mengalami fragmentasi yang signifikan.

## 3. Closeness Centrality

    Interpretasi: Closeness centrality mengukur kedekatan rata-rata node dengan semua node lainnya, menandakan efisiensi dalam mencapai node lain.
    
    Temuan: Node dengan closeness centrality tertinggi (misalnya, 1308 dan 2594) relatif lebih dekat dengan node lain, menjadikannya efisien dalam mendistribusikan daya.

## 4. Eigenvector Centrality

    Interpretasi: Node dengan eigenvector centrality tinggi memiliki koneksi ke node lain yang juga sangat terhubung, menunjukkan pengaruh global.
    
    Temuan: Node 4381 dan 4345 adalah node yang sangat terhubung dengan node penting lainnya. Hal ini mengindikasikan bahwa node ini memiliki pengaruh signifikan dalam jaringan.

## 5. PageRank

    Interpretasi: PageRank mirip dengan eigenvector centrality tetapi lebih memperhatikan struktur lokal. Node dengan PageRank tinggi memiliki pengaruh karena koneksi langsung maupun tidak langsungnya.
    
    Temuan: Node seperti 4458, 831, dan 3468 memiliki peran penting dalam keterhubungan jaringan secara lokal, meskipun mungkin tidak selalu menjadi node utama secara global.

```python
import matplotlib.pyplot as plt

# Average Degree
avg_degree = sum(dict(G.degree()).values()) / G.number_of_nodes()
print("Average Degree:", avg_degree)

# Degree Distribution
degree_sequence = [d for n, d in G.degree()]
plt.hist(degree_sequence, bins=range(min(degree_sequence), max(degree_sequence) + 1), alpha=0.75, color='blue', edgecolor='black')
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
# Average Clustering Coefficient
avg_clustering = nx.average_clustering(G)
print("Average Clustering Coefficient:", avg_clustering)

# Compute clustering coefficients for all nodes
clustering_coefficients = nx.clustering(G)
 
# Sort nodes by clustering coefficient in descending order
sorted_clustering = sorted(clustering_coefficients.items(), key=lambda x: x[1], reverse=True)

# Get the top 10 nodes
top_10 = sorted_clustering[:10]

# Print the top 10 nodes and their clustering coefficients
print("Top 10 Clustering Coefficients:")
for node, coefficient in top_10:
    print(f"Node: {node}, Clustering Coefficient: {coefficient}")

down_10 = sorted_clustering[-9:]

# Print the top 10 nodes and their clustering coefficients
print("Bottom 10 Clustering Coefficients:")
for node, coefficient in down_10:
    print(f"Node: {node}, Clustering Coefficient: {coefficient}")
```

Clustering Coefficient

    Average Clustering Coefficient (0.08): Tingkat clustering yang rendah menunjukkan konektivitas lokal yang minimal, yang umum dalam jaringan distribusi daya, di mana prioritasnya adalah distribusi daya yang terstruktur, bukan redundansi.

```python
# Diameter
if nx.is_connected(G):  # Diameter is only defined for connected graphs
    diameter = nx.diameter(G)
    print("Diameter:", diameter)
else:
    print("Graph is not connected; diameter undefined.")

# Average Path Length
if nx.is_connected(G):
    avg_path_length = nx.average_shortest_path_length(G)
    print("Average Path Length:", avg_path_length)
else:
    print("Graph is not connected; average path length undefined.")
```

Struktur Global Jaringan

    Average Degree (2.67): Jaringan ini sangat sparsely connected, khas jaringan fisik seperti power grid yang tidak membutuhkan redundansi tinggi seperti jaringan sosial.
    Diameter (46): Diameter tinggi menunjukkan skalabilitas jaringan yang luas, tetapi juga menunjukkan bahwa propagasi daya mungkin lambat di jaringan ini.
    Average Path Length (18.99): Jarak rata-rata yang tinggi menunjukkan konektivitas yang efisien tetapi mungkin memerlukan peningkatan jika kegagalan node terjadi.

```python
import numpy as np

# Degree distribution
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_counts = np.bincount(degree_sequence)
degree = np.arange(len(degree_counts))
degree = degree[degree_counts > 0]
degree_counts = degree_counts[degree_counts > 0]

# Log-log plot
log_degree = np.log(degree)
log_counts = np.log(degree_counts)

# Linear fit to find scaling exponent
fit = np.polyfit(log_degree, log_counts, 1)
scaling_exponent = -fit[0]  # Negative slope of the line
print("Scaling Exponent:", scaling_exponent)

# Plot
plt.scatter(log_degree, log_counts, label="Data")
plt.plot(log_degree, np.polyval(fit, log_degree), label="Fit", color="red")
plt.title("Scaling Exponent (Log-Log)")
plt.xlabel("log(Degree)")
plt.ylabel("log(Frequency)")
plt.legend()
plt.show()
```

Scaling Exponent (2.84)

    Scaling exponent mendekati 3 menunjukkan bahwa jaringan ini memiliki struktur skala small-world (mirip jaringan skala-bebas). Sebagian besar node memiliki koneksi kecil, tetapi beberapa node berfungsi sebagai hub besar dengan konektivitas tinggi.

Jaringan memiliki komunitas yang sangat jelas dan terpisah dengan interaksi internal yang kuat.



ada beberapa node dengan degree sangat tinggi (hub/penyambung utama) dan banyak node dengan degree rendah.

```python
# Assortativity Coefficient
assortativity = nx.degree_assortativity_coefficient(G)
print("Assortativity Coefficient:", assortativity)
```

Assortativity

    Koefisien Asortativitas (0.0034): Nilai mendekati nol menunjukkan bahwa jaringan ini non-asortatif, artinya node dengan banyak koneksi (hub) tidak cenderung terhubung ke hub lain, tetapi lebih cenderung terhubung ke node dengan konektivitas rendah. Ini adalah karakteristik jaringan distribusi daya untuk meningkatkan efisiensi.

Node terhubung secara acak berdasarkan degree tanpa pola jelas (non-preferential).

```python
from community import community_louvain

# Louvain Modularity
partition = community_louvain.best_partition(G)
modularity = community_louvain.modularity(partition, G)
print("Modularity:", modularity)
```

Modularitas


    Modularity (0.937): Nilai modularitas yang sangat tinggi menunjukkan adanya komunitas terpisah yang jelas dalam jaringan. Hal ini sesuai dengan jaringan power grid yang cenderung memiliki kelompok node yang mencerminkan area geografis atau fungsi tertentu.

# Kesimpulan dan Keunikan

Jaringan Western States Power Grid memiliki beberapa karakteristik unik:

    Efisiensi Lokal: Hub-hub utama (dengan centrality tinggi) memainkan peran penting dalam distribusi daya.
    Kerentanan Strategis: Node dengan betweenness dan eigenvector centrality tinggi rentan terhadap kegagalan yang dapat menyebabkan gangguan besar pada jaringan.
    Komunitas Modular: Tingginya modularitas menunjukkan pembagian alami jaringan sesuai wilayah atau fungsi, mendukung stabilitas operasi lokal.
    Struktur Small-World: Struktur ini memastikan efisiensi global meski jarak rata-rata antar node cukup tinggi.

Implikasi

Jaringan ini stabil secara geografis dan hierarkis, tetapi untuk meningkatkan ketahanan, redundansi pada node strategis atau penambahan koneksi di wilayah yang sparsely connected dapat membantu untuk membuat jaringan power grid menjadi lebih robust.
